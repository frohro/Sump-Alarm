# Sump Pump Alarm
It would be a disaster if the sump overflows in our basement, so I made an alarm using an ESP8266 and an H-Bridge and speaker.  This is actually the second incarnation of the sump alarm.  The first relied on my server being up and it isn't these days.  This one uses Telegram and Pushover.  The code was mostly written by Claude 3.7 Sonnet.  It uses MicroPython.  Here is the pinout of the NodeMCU V3 by LoLin I used.



The pins I used are:
- Water sensor is on GPIO 5 (D1 on NodeMCU)
- LED indicator is on GPIO 2 (D4 on NodeMCU, the blue LED)
- Speaker H-bridge is driven by:
  - GPIO 12 (D6) for IN_A
  - GPIO 14 (D5) for IN_B

I used an L9110S H-Bridge.  Instead of a motor, the H-Bridge drives a speaker that should make a loud noise in the pump room.  It also sends notifications via Telegram and Pushover.  The sensing is done looking for conductivity of the sewage.  An internal pull-up resistor has GPIO 5 normally HIGH, but if a flood occurs, it should go LOW.
